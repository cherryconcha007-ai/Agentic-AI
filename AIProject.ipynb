```python
import sys
if 'streamlit' not in sys.modules:
    !pip install streamlit
import streamlit as st
import pandas as pd
import plotly.express as px

# --- MOCK DATA GENERATION (Replace with your Agent Output) ---
data = {
    "Company": ["NVIDIA", "Vertiv", "Arista Networks", "Eaton", "Supermicro"],
    "Layer": ["Compute", "Cooling", "Networking", "Power", "Compute"],
    "Moat": [5.0, 4.2, 4.5, 3.8, 2.5],
    "Margin": [5.0, 3.5, 4.0, 3.0, 2.0],
    "Growth": [0.45, 0.30, 0.25, 0.20, 0.50],
    "Risk": [0.95, 0.90, 0.95, 1.0, 0.60]
}
df = pd.DataFrame(data)
df['TAFGS'] = (df['Moat'] * df['Margin']) * df['Growth'] * df['Risk']
df = df.sort_values(by="TAFGS", ascending=False)

# --- STREAMLIT UI ---
st.set_page_config(page_title="AI Factory Growth Equity Tracker", layout="wide")
st.title("🚀 AI Factory Growth Equity Identification")
st.markdown("### Agentic Ranking: Top 20 Growth Scoring")

# Sidebar Filters
st.sidebar.header("Filters")
selected_layer = st.sidebar.multiselect("AI Stack Layer", options=df['Layer'].unique(), default=df['Layer'].unique())
filtered_df = df[df['Layer'].isin(selected_layer)]

# Main Metrics
col1, col2, col3 = st.columns(3)
col1.metric("Top Pick", filtered_df.iloc[0]['Company'], f"{filtered_df.iloc[0]['TAFGS']:.2f} Score")
col2.metric("Avg. Moat Score", f"{filtered_df['Moat'].mean():.2f}")
col3.metric("Avg. Growth (3Y)", f"{filtered_df['Growth'].mean()*100:.1f}%")

# --- VISUALIZATIONS ---
st.divider()

c1, c2 = st.columns([2, 1])

with c1:
    st.subheader("The TAFGS Leaderboard")
    # Interactive Bar Chart
    fig = px.bar(filtered_df, x='Company', y='TAFGS', color='Layer',
                 hover_data=['Moat', 'Margin', 'Growth'],
                 title="Total AI Factory Growth Score (Risk Adjusted)")
    st.plotly_chart(fig, width='stretch')

with c2:
    st.subheader("Moat vs. Margin Matrix")
    # Scatter plot to identify "Quality" outliers
    fig_scatter = px.scatter(filtered_df, x='Moat', y='Margin', size='Growth',
                             color='Company', hover_name='Company',
                             title="High Moat / High Margin Leaders")
    st.plotly_chart(fig_scatter, width='stretch')

# Data Table
st.subheader("Raw Agentic Output")
st.dataframe(filtered_df.style.background_gradient(subset=['TAFGS', 'Moat'], cmap='BuGn'))
```
